### 그래프에 옵션 추가하기
    기본 line plot을 조금 더 색다르게 꾸밀 수 있는 스타일을 지금까지 알아봤다면 
    지금부터는 plt 의 다른 함수를 사용하여 보는 사람의 이해를 도울 수 있는 여러 기능들을 알아보겠습니다.
    
    figure : 그래프가 그려지는 캔버스 설정
    title : 그래프 제목
    xlabel, ylabel : 축 설명
    legend : 범례이름
    xlim, ylim : 축 범위
    xticks, yticks : 축 구간 내 필요한 구간만 추리기

```python
# 그래프 크기조정
plt.figure(figsize=(10, 8))
plt.plot([0, 1, 2, 3, 4, 5],[0, 1, 4, 9, 16, 25], 'g:')
plt.show()
```

```python
# 그래프 타이틀 지정
plt.figure(figsize=(10, 7))
plt.title('x square')
plt.plot(x, y,'g--')
plt.show()
```

```python
# 축 타이틀 지정
plt.figure(figsize=(10, 7))
plt.title('제곱그래프')
plt.plot([0, 1, 2, 3, 4, 5],[0, 1, 4, 9, 16, 25], 'g:')
plt.ylabel('x^2')
plt.xlabel('x')
plt.show()
```

```python
# 범례지정
plt.figure(figsize=(10, 7))
plt.title('제곱그래프')
#$표시를 텍스트안에 넣으면 수학식처럼 표현이 됨.
plt.plot([0, 1, 2, 3, 4, 5],[0, 1, 4, 9, 16, 25], 'm:', label='$x^9$')
plt.ylabel('x^2')
plt.xlabel('x')
#legend는 범례, label과 세트로 쓰임. 하나라도 없으면 실행이 안됨.
plt.legend(loc='best')
plt.show()
```

```python
# 축 범위 설정
plt.figure(figsize=(10, 7))
plt.title('제곱그래프')
plt.plot([0, 1, 2, 3, 4, 5],[0, 1, 4, 9, 16, 25], 'g:', label='$x^2$')
plt.ylabel('x^2')
plt.xlabel('x')
#xlim, ylim은 제한하는 것.
plt.xlim([2, 7])
plt.ylim([9, 50])
plt.legend(loc='best')
plt.show()
```

```python
# 축에 표시 될 내용만 사용하기
plt.figure(figsize=(10, 7))
plt.title('제곱그래프')
plt.plot([0, 1, 2, 3, 4, 5],[0, 1, 4, 9, 16, 25], 'y--', label='$x^2$')
plt.ylabel('x^2')
plt.xlabel('x')
plt.xlim([3, 5])
plt.ylim([9, 25])
plt.xticks([3, 4, 5])
plt.yticks([10, 15, 30])
plt.legend(loc='best')
plt.show()
```

```python
# - 부호 나오게 하는 법.
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 여러 그래프 한꺼번에 그리기
plt.figure(figsize=(10, 7))
plt.title('$x^2$ and $x^3$')
x = np.linspace(-10, 10, 300)
plt.plot(x, x*x, label='$x^2$')
plt.plot(x, x*x*x, 'r--', label='$x^3$')
plt.plot(x, np.sin(x), 'g:', label='$sin(x)$')
plt.plot(x, np.cos(x), 'k:', label='$cos(x)$')
plt.ylabel('$x^2$ and $x^3$')
plt.xlabel('x')
# plt.xlim(-3, 3)
# plt.ylim(-1, 1)
# plt.xticks([-2,0,2])
# plt.yticks([-2,-1,1])
plt.legend(loc='best')
plt.show()
```

### subplot을 사용하여 여러 그래프 한번에 그리기

```python
# 2 x 1 figure 매트릭스에 그래프 그리기

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.figure(figsize=(8, 4))

ax1 = plt.subplot(2, 1, 1) # 2:행, 1, 열, 1:순번
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout() # 서로 다른 그래프 사이의 여백줄이기
plt.show()
```

```python
# 2 x 2 figure 매트릭스에 그래프 그리기

plt.figure(figsize=(12, 8))
np.random.seed(0)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```
