**8.(option) 7번 까지 task를 완수하셨다면 전체데이터셋을 불러들여서 7번 테스트와 같이 문현동 공적마스트 판매점을 확인해봅시다.**

```python
# CODE
df = pd.DataFrame()
for i in range(1, 7):
    url = f'https://8oi9s0nnth.apigw.ntruss.com/corona19-masks/v1/stores/json?page={i}&perPage=5000.'
    #요청하고
    response = requests.get(url).text
    #데이터 받고
    data = json.loads(response)
    df_temp = pd.DataFrame(data['storeInfos'])    # 전달받은 데이터
    # 데이터프레임 병합
    df = pd.concat([df, df_temp], ignore_index=True)
    
```
