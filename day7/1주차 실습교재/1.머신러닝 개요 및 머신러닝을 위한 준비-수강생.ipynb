```python
## 분류/예측 모델 개발 가이드라인 실습
# 1. 데이터 불러오기 6999
import seaborn as sns
iris = sns.load_dataset('iris') # seaborn에 있는 iris 데이터 가져오기
X_iris = iris.drop('species', axis=1) # species 열을 제거한 데이터 프레임
y_iris = iris['species'] # species 열만 가져옴

# 2. 데이터 분리하기 
from sklearn.model_selection import train_test_split
# To do
Train_X, Test_X, Train_Y, Test_Y = train_test_split(X_iris, y_iris, test_size=0.2, random_state=2019)

# 3. 모델 클래스 선택
from sklearn.naive_bayes import GaussianNB 

# 4. 모델 인스턴스화 
model = GaussianNB() 

# 5. 모델을 데이터에 적합     
# to do  
model.fit(Train_X, Train_Y)

# 6. 새 데이터에 대해 예측
# to do 
predicted = model.predict(Test_X)

# 7. 모델 평가

from sklearn.metrics import accuracy_score 
# to do               
accuracy_score(Test_Y, predicted)
```

```python
##k겹 교차 검증 맛보기
import seaborn as sns
iris = sns.load_dataset('iris')
X_iris = iris.drop('species', axis=1)
y_iris = iris['species']

# KNN 모델 인스턴스화
from sklearn.neighbors import KNeighborsClassifier
model1 = KNeighborsClassifier(n_neighbors=1)
model2 = KNeighborsClassifier(n_neighbors=3)


# 5-겹 교차 검증
from sklearn.model_selection import cross_val_score
scores1 = cross_val_score(model1, X_iris, y_iris, cv=5, scoring = 'accuracy')
scores2 = cross_val_score(model2, X_iris, y_iris, cv=5, scoring = 'accuracy')

print(scores1.mean())
print(scores2.mean())
```
